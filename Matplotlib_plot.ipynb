##***CRO***

```python


# data
frequency = 600  # dynamic Hz
peak_voltage = 5  # dynamic Volts

period = 1 / frequency  # seconds
time = np.linspace(0, 2 * period, 1555)  # two cycles for better visualization

# Calculate the waveform
voltage = peak_voltage * np.sin(2 * np.pi * frequency * time)

# Plot the waveform
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(time, voltage, label='Voltage (V)')

ax.set_title('Voltage Waveform')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Voltage (V)')

ax.axhline(y=0, color='k', linestyle='--', linewidth=0.8)
ax.axhline(y=peak_voltage, color='r', linestyle='--', linewidth=0.8, label='Peak Voltage (5V)')
ax.axhline(y=-peak_voltage, color='r', linestyle='--', linewidth=0.8)
ax.axhline(y=peak_voltage/np.sqrt(2), color='g', linestyle='--', linewidth=0.8, label='RMS Voltage (~3.54V)')
ax.axhline(y=-peak_voltage/np.sqrt(2), color='g', linestyle='--', linewidth=0.8)

ax.legend()
ax.set_xticks([])
ax.set_yticks([])

# show plot
plt.show()
```

# ***Error Bar***

```python
x = [40, 20, 10, 5, 2.5]

y = [50.5, 24.4, 11.7, 5.55, 2.9]

# find slop from x and y
m = np.polyfit(x, y, 1)[0]
# polyfit function used for find slop

#  find c
c = np.polyfit(x, y, 1)[1]

# find y_the
# y_the = m * x + c
y_the = np.polyval([m, c], x)

# find error
error = np.std(y) / np.sqrt(len(x))

# plot
fig, ax = plt.subplots(2, 1)

fig.set_size_inches(16, 8.27)

ax[0].plot(x, y, 'o-', label='Experimental Data')
ax[0].plot(x, y_the, 'r-', label='Pridicted Data')
ax[0].legend()
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_ylim(0, 10)
ax[0].set_xlim(2, 10)

ax[1].errorbar(x, y, yerr=error, fmt='o-', label='Experimental Data with error', linewidth=2, capsize=6)
ax[1].legend()
ax[1].set_xlabel('x')
ax[1].set_ylabel('y' + 'Error')
ax[1].set_ylim(0, 60)
ax[1].set_xlim(0, 60)


plt.show()
```

# ***Find Density***

```python
#for finding the population density at 2023
area=df['area (km²)'].head(10)
pop=df['2023 population'].head(10)
density=pop/area
#find correlation
corr=np.corrcoef(density,area)
print(corr)
n_df=pd.DataFrame({'country':df['country'].head(10),'density':density,'area':area,'pop':pop})
print(n_df)
#plot
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.hist(n_df['country'].head(10),weights=n_df['density'].head(10),linewidth=1.5,edgecolor='white')
#Labelling
ax.set_title('World Population Density')
ax.set_xlabel('country')
ax.set_ylabel('density (km²)')
#ticks and tick label
ax.set_xticks(range(len(n_df['country'].head(10))))
ax.set_xticklabels(n_df['country'].head(10),rotation=90,color='red')
#styles
plt.style.use('dark_background')
#grid
ax.grid(False)
#legend
ax.legend()
#show
plt.show()
```
