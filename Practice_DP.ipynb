```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math as m


```

```python
#Read url
df=pd.read_csv('https://raw.githubusercontent.com/jsdhami/Python-For-Research/main/Data/seattle-weather.csv')
df=df.tail(20)
df.head()
```

```python
#Select data
x=df['date']
y=df['precipitation']
z=df['temp_max']
a=df['temp_min']
b=df['wind']



```

# Line plot

```python
#Line plot for precipitation
#create figure
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
#PLots
ax.plot(x,y,label='Precipitation')
#labelling
ax.set_xlabel('Date',color='red')
ax.set_ylabel('Precipitation',color='red')
ax.set_title('Precipitation in Seattle',color='red')

#set ticks and ticks labels
ax.set_xticks(x[::2])
ax.set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)


#styles
plt.style.use('dark_background')
plt.grid(False)
plt.legend()
plt.show()
```

```python
#line plot for temperature
#create figure
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
#plots
ax.plot(x,z,label='Max Temp')
ax.plot(x,a,label='Min Temp')

#labelling
ax.set_xlabel('Date',color='blue')
ax.set_ylabel('Temperature',color='blue')
ax.set_title('Temperature in Seattle',color='purple')

#set ticks and ticks labels
ax.set_xticks(x[::2])
ax.set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)

#styles
plt.style.use('dark_background')
plt.grid(False)
plt.legend()
plt.show()
```

```python
#line plot for wind
#create figure
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
#plots
ax.plot(x,b,label='Wind')

#labelling
ax.set_xlabel('Date',color='green')
ax.set_ylabel('Wind',color='green')
ax.set_title('Wind in Seattle',color='green')

#set ticks and ticks labels
ax.set_xticks(x[::3])
ax.set_xticklabels(x[::3],rotation=85,color='green',fontsize=10)

#styles
plt.style.use('dark_background')
plt.grid(False)
plt.legend()
plt.show()
```

## 2x2 plots

```python
# 2x2 plots
#create figure
fig,ax=plt.subplots(2,2)
fig.set_size_inches(15,10)
#plots
#plot[0,0]
ax[0,0].plot(x,y,label='Precipitation')
ax[0,0].set_title('Precipitation in Seattle')
ax[0,0].set_xlabel('Date',color='blue')
ax[0,0].set_ylabel('Precipitation',color='red')
#plots[0,1]
ax[0,1].plot(x,b,label='Wind')
ax[0,1].set_title('Wind in Seattle')
ax[0,1].set_xlabel('Date',color='green')
ax[0,1].set_ylabel('Wind',color='red')
#plots[1,0]
ax[1,0].plot(x,z,label='Max Temp')
ax[1,0].set_title('Max Temp in Seattle')
ax[1,0].set_xlabel('Date',color='green')
ax[1,0].set_ylabel('Temperature',color='orange')
#plots[1,1]
ax[1,1].plot(x,a,label='Min Temp')
ax[1,1].set_title('Min Temp in Seattle')
ax[1,1].set_xlabel('Date',color='green')
ax[1,1].set_ylabel('Temperature',color='orange')
#styles
plt.style.use('dark_background')
#ticks and ticks labels
ax[0,0].set_xticks(x[::2])
ax[0,0].set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)
ax[0,1].set_xticks(x[::2])
ax[0,1].set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)
ax[1,0].set_xticks(x[::2])
ax[1,0].set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)
ax[1,1].set_xticks(x[::2])
ax[1,1].set_xticklabels(x[::2],rotation=85,color='green',fontsize=10)
#grid
plt.grid(False)
#legend
plt.legend()
plt.tight_layout()
plt.show
```
