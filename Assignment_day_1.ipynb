```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math as m
```

```python
#Read url
df=pd.read_csv('https://raw.githubusercontent.com/jsdhami/Python-For-Research/main/Data/hourly_ktm_weather_data.csv')
df=df.tail(20)
df.head()
```

```python
# create data frame
df = pd.DataFrame({'date': df['date'], 'temp': df['temperature_2m'], 'ppt': df['precipitation'], 'windspeed': df['wind_speed_10m']})

df.head()
```

```python
#select data
x = df['date']
y = df['temp']
z = df['ppt']
w = df['windspeed']
```

```python
# stem plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 6)
ax.stem(x,y, label="Temp.")
#labelling
ax.set_xlabel('Date')
ax.set_ylabel('Temp,PPT,Wind Speed')


# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)



# set limit on x-axis
ax.set_xlim(0, 10)

# set title
ax.set_title('Temperature')
#style
plt.style.use('dark_background')

plt.grid(False)
plt.show()
```

```python
#Line plot
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.plot(x,y, label="Temp.",color='blue')
ax.plot(x,z, label="PPT.",color='red')
ax.plot(x,w, label="Wind Speed",color='green')

#labelling
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15,color='green')
ax.set_title('weather condition of kathmandu 2024',fontsize=15,color='red')


#set ticks and ticks label
ax.set_xticks(x[::2])
ax.set_xticklabels(x[::2],rotation=90,fontsize=8)

#style
plt.style.use('dark_background')

#legends
ax.legend()

#grid
plt.grid(False)

#show graph
plt.show()
```

```python
#scatter plot
#create fig
fig,ax=plt.subplots()
fig.set_size_inches(15,5)

#plot
ax.scatter(x,y, label="Temp.",color='blue')
ax.scatter(x,z, label="PPT.",color='red')
ax.scatter(x,w, label="Wind Speed",color='green')
#labelling
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15)
ax.set_title('Temperature',fontsize=15)

#ticks
ax.set_xticks(x[::2])
ax.set_xticklabels(x[::2],rotation=79,fontsize=8)

#style
plt.style.use('dark_background')

#legends
ax.legend()

#grid
plt.grid(False)

#show graph
plt.show()
```

```python
#Stem plot
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.stem(x,y, label="Temp.",markerfmt='blue')
ax.stem(x,z, label="PPT.",markerfmt='red')
ax.stem(x,w, label="Wind Speed",markerfmt='green')

#labelling
ax.set_xlabel('Date',fontsize=15,color='blue')
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15,color='green')
ax.set_title('KTM weather Data',fontsize=15,color='red')

#ticks
ax.set_xticks(x[::2])
ax.set_xticklabels(x[::2],rotation=90,fontsize=8)

#style
plt.style.use('dark_background')

#legends
ax.legend()
#grid
plt.grid(False)

#show graph
plt.show()
```

```python
#Bar plot for temp
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.bar(x,y, label="Temp.",color='blue')

#labelling
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temp',fontsize=15)
ax.set_title('Temperature',fontsize=15)

#ticks
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1],rotation=90,fontsize=8)

#style
plt.style.use('dark_background')
ax.set_facecolor('violet')

#legends
ax.legend()

#grid
plt.grid(False)

#show graph
plt.show()
```

```python
#stack plot.
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.stackplot(x,y,z,w, labels=['Temp.','PPT.','Wind Speed'],colors=['blue','red','green'])

#labelling
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15)
ax.set_title('Weather condition of ktm valley 2024',fontsize=15)

#ticks
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1],rotation=90,fontsize=8)

#style
plt.style.use('dark_background')

#legends
ax.legend()

#grid
plt.grid(False)

#show graph
plt.show
```

```python
#stairs plotting
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.stairs(y,label='Temp.',color='green')
ax.stairs(z,label='PPT.',color='blue')
ax.stairs(w,label='Wind Speed',color='grey')

#labelling
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15)
ax.set_title('Weather condition of ktm valley 2024',fontsize=15)

#ticks
ax.set_xticklabels(x[::2],rotation=90,fontsize=8)

#style
plt.style.use('dark_background')

#legends
ax.legend()

#show
plt.show()
```

```python
#fill between plot
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.fill_between(x,y,z,label='Temp.',color='green')
ax.fill_between(x,z,w,label='PPT.',color='blue' )


#labelling
ax.set_xlabel('Date',fontsize=15,color='blue')
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15,color='red')
ax.set_title('Weather condition of ktm valley 2024',fontsize=15,color='pink')
#ticks
ax.set_xticklabels(x[::1],rotation=90)

#style
plt.style.use('dark_background')

#legends
ax.legend()
#grod
plt.grid(False)

#show
plt.show()
```

```python
#hist plot
fig,ax=plt.subplots()
fig.set_size_inches(15,5)
ax.hist(y,label='Temp.',color='green')
ax.hist(z,label='PPT.',color='blue')
ax.hist(w,label='Wind Speed',color='pink')

#labelling
ax.set_xlabel('Date',fontsize=15,color='purple')
ax.set_ylabel('Temp,PPT,Wind Speed',fontsize=15,color='green')
ax.set_title('Weather condition of ktm valley 2024',fontsize=15,color='yellow')

#ticks
ax.set_xticklabels(x[::1],rotation=88)


#style
plt.style.use('dark_background')

#legends
ax.legend()

#grid
plt.grid(True)

#show
plt.show()
```

```python
#create one 2x2 axis
fig,ax=plt.subplots(2,2)
fig.set_size_inches(20,10)
#plot[0,0]
ax[0,0].plot(x,y, label="Temp.",color='blue')
ax[0,0].set_title('Temp')
ax[0,0].set_xlabel('Date')
ax[0,0].set_ylabel('Temp')
#plot[0,1]
ax[0,1].plot(x,z, label="PPT.",color='red')
ax[0,1].set_title('PPT')
ax[0,1].set_xlabel('Date')
ax[0,1].set_ylabel('PPT')

#plot[1,0]
ax[1,0].plot(x,w, label="Wind Speed",color='green')
ax[1,0].set_title('Wind Speed')
ax[1,0].set_xlabel('Date')
ax[1,0].set_ylabel('Wind Speed')

#plot[1,1]
ax[1,1].plot(x,y, label="Temp.",color='blue')
ax[1,1].set_title('Temp')
ax[1,1].set_xlabel('Date')
ax[1,1].set_ylabel('Temp')

#style
plt.style.use('dark_background')

#ticks
ax[0,0].set_xticks(x[::2])
ax[0,0].set_xticklabels(x[::2],rotation=90)
ax[0,1].set_xticks(x[::2])
ax[0,1].set_xticklabels(x[::2],rotation=90)
ax[1,0].set_xticks(x[::2])
ax[1,0].set_xticklabels(x[::2],rotation=90)
ax[1,1].set_xticks(x[::2])
ax[1,1].set_xticklabels(x[::2],rotation=90)


#legend
plt.legend()
plt.tight_layout()
#show
plt.show()
```
